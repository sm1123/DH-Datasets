# <font color="limegreen"> Line Plot

I want to investigate if the drivers that won the most "driver of the day awards", resulted in more race wins.

```python
df.groupby("dotd")["wins"].sum().sort_values().plot.line(color = "salmon",
                                                           marker = "o",
                                                           title = "Number of Wins in relation to DOTD awards ",
                                                           xlabel = "Number of Driver of the day Awards",
                                                           ylabel = "Number of Wins",
                                                           figsize = (10,7))
```

# <font color="limegreen"> Analysis of Line Plot:

I wanted to check if the number of driver of the day awards were typically awarded to the driver who won the race. I grouped by "dotd" for the x-axis and "wins" for the y-axis in order to see a trend. From this analysis I can conclude that drivers with the most "driver of the day" awards have the most wins. What I can read from the line plot is that drivers that win the best "driver of the day" for that particular race, tend to have more and more wins. As a fan of Formula One, it's not always the case that the driver that wins the race earns the award for "driver of the day". However, this graph shows that last year, drivers with more of these awards would have more race wins. Drivers who win more of the best "driver of the day" awards will result in more race wins. The driver who had the most "driver of the day" awards with 5, won 15 races that season. On the other hand, the driver with 0 "driver of the day" awards had only 1 race win. This line plot shows that the drivers that win more "driver of the day awards" are more likely to win the race. I was a little bit surprised by these findings because oftentimes, the driver who wins the "driver of the day" award doesn't always end up winning the race but they might have had an excellent race. This means the driver who won the most awards and the most races was consistently the best during the season.

# <font color="limegreen"> Area Plot

I want to investigate the metascore and Adjusted Score ratings of the Box Office movies over time.

```python
df.groupby("Year").sum().tail(60).plot.area(y = ["metascore", "Adjusted Score"],
             figsize = (6,5), alpha = 0.3,
             xlabel = "Year",
             ylabel = "Meta Score and Adjusted Score over time",
             title = "Area plot of metascore and Adjusted Score of Box Office movies since 1970",
             stacked = False, color = ["orange", "springgreen"])
```

# <font color="limegreen"> Area Plot Analysis:

I started off by grouping by "Year" so I could track the progression or regression of the "metascore" and "Adjusted Score" ratings over time. Straight away I can see that the adjusted score is always higher than the metascore. The metascore is the score for the box office movies that are rated by movie critics and reviwers that are usually very hard with their ratings for movies. It's not surprising to see the adjusted score of these movies higher than the metascore as a result. I'm assuming the adjusted score is the rating for the movies after they've been released and after they got feedback from the movies to see how well they were doing, and then they adjusted it accordingly. Both scores are increasing from 1970 to 2020 which is not a surprise. Movies and in particular box office movies are getting more popular in the world as time goes on and more money is being spent on the production of movies so I would expect to see an increase in their ratings in the 21st century compared to the 20th century.
